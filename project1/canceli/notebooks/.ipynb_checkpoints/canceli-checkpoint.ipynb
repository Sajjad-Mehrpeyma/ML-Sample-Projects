    در این سوال شما می‌توانید از هر تکنیک پیش‌پردازش/مهندسی ویژگی که در فصل‌های گذشته آموختید، استفاده کنید.
    <br>
    تکنیک‌هایی که استفاده می‌کنید به شکل مستقیم مورد ارزیابی توسط سامانه داوری قرار <b>نمی‌گیرند.</b> بلکه همه آن‌ها در دقت مدل شما تاثیر خواهند گذاشت؛ بنابراین هر چه پیش‌پردازش/مهندسی ویژگی بهتری انجام دهید تا دقت مدل بهبود پیدا کند، امتیاز بیشتری از این سوال کسب خواهید کرد.
</font>
</p>

```python
y = train_data['Cancel']

# dropping some null columns
df = train_data.drop(['CancelTime', 'UserID', 'HashPassportNumber_p', 'HashPassportNumber_p', 'HashEmail', 'Cancel'], axis=1)
```

```python
df.isna().sum()
```

### now lets make a little model to fill null values

```python
df.describe()
```

```python
df.head()
```

```python
df = df.drop(['Male', 'BillID', 'TicketID', 'VehicleType', 'BuyerMobile', 'NationalCode'], axis=1)
```

```python
df['month_created'] = df['Created'].apply(lambda x : x.split('-')[1])
df['month_departure'] = df['DepartureTime'].apply(lambda x : x.split('-')[1])
```

```python
type(df["month_created"][0])
```

```python
df.drop(['Created', 'DepartureTime'], axis=1, inplace=True)
```

```python
df.drop(['From', 'To'], axis=1, inplace=True)
```

```python
df.head()
```

```python
temp = pd.get_dummies(df)
```

```python
temp
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(temp, y, test_size=0.25)
```
