## K Means Cluster Creation

<div style="direction:rtl">
                                     (  )  .

```python
y = df['Private']
X = df.drop('Private', axis=1)
```

```python
y_binary = y.map(lambda x : 1 if x=='Yes' else 0)
```

```python
from sklearn.cluster import KMeans

model = KMeans(n_clusters=2, n_init='auto')
model.fit(X)
pred = model.predict(X)
```

<div style="direction:rtl">
   :

```python
model.cluster_centers_
```

## Evaluation

<div style="direction:rtl">
                                       .

<div style="direction:rtl">
        :

<div style="direction:rtl">
           0  1          Private     0  1                .

```python
from sklearn.metrics import silhouette_score

silhouette_score(np.array(y_binary).reshape(-1, 1), pred)
```

<div style="direction:rtl">
           score                                 .

```python
from sklearn.metrics import classification_report

print(classification_report(np.array(y_binary).reshape(-1, 1), pred))
```

<div style="direction:rtl">
                                         .              .

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
scaled_df = ss.fit_transform(X)
```

```python
model2 = KMeans(n_clusters=2)
model2.fit(scaled_df)
pred2 = model2.predict(scaled_df)
```

```python
from sklearn.metrics import classification_report

print(classification_report(np.array(y_binary).reshape(-1, 1), pred2))
```

<div style="direction:rtl">
                          10  65  .                                

```python
scores = []
for i in range(2, 11):
	model_temp = KMeans(n_clusters=i, n_init='auto')
	model_temp.fit(scaled_df)
	pred_temp = model_temp.predict(scaled_df)
    
	score = silhouette_score(np.array(pred_temp).reshape(-1, 1), y_binary)
	scores.append(score)
```

```python
sns.lineplot(scores)

# ?????
```
